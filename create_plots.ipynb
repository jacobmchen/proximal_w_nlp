```python
from synthetic_dgp import *
import pandas as pd
import matplotlib.pyplot as plt
```

### True ACE = 0

```python
backdoor_distribution_afib = run_backdoor_adjustment(oracle='afib', causal_effect=0)[1]
```

```python
hypothesis_test_distribution_afib = run_causal_null_hypothesis(oracle='afib', causal_effect=0)
```

```python
plot_data = pd.DataFrame({'Backdoor': backdoor_distribution_afib[2], 
                          'Hypothesis Test': hypothesis_test_distribution_afib[2]})

fig, ax = plt.subplots()

plot_data.boxplot(whis=(2.5, 97.5), showfliers=False, ax=ax)
plt.title('Atrial Fibrillation')
plt.ylim([-0.05, 0.40])

plt.show()
```
